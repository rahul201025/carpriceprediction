```python
# here carname is not required or play importanant rolw in prediction selling price
final_dataset= df[[ 'Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_dataset
```

```python
final_dataset['Current_Year']=2023
final_dataset.head()
```

```python
#here from year column we can not find how old it is so to calculte its depriciation we nned to 
#   subtract year value from current value & drop Year & current year after getting no of old years values column

final_dataset['No_Years']=final_dataset['Current_Year']-final_dataset['Year']
final_dataset
```

```python
final_dataset.drop(["Year","Current_Year"],axis=1,inplace=True)
```

```python
final_dataset
```

```python
final_dataset=pd.get_dummies(final_dataset,drop_first=True)
final_dataset
```

```python
#find out correlation using .corr

final_dataset.corr()
```

```python
corrmat=final_dataset.corr()
top_corr_features=corrmat.index
plt.figure(figsize=(7,7))
g=sns.heatmap(final_dataset[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
#seprate dependent and independent features

X= final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
#featue importance

from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,y)
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series (model. feature_importances_, index=X.columns)
feat_importances.nlargest (5). plot (kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
# Hyper Parameters used 

 n_estimators=100, (10 to 100)
    *,
    criterion='squared_error',     criterion : {"squared_error", "absolute_error", "poisson"}, default="squared_error"
    
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features='auto',
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    bootstrap=True,
    oob_score=False,
    n_jobs=None,
    random_state=None,
    verbose=0,
    warm_start=False,
    ccp_alpha=0.0,
    max_samples=None,
```

```python
###Hypeparameters

import numpy as np

n_estimators = [int(x) for x in np.linspace (start = 100, stop = 1200, num = 12)] 
print(n_estimators)
```

```python
#Randomized Search CV

# Number of trees in random forest

n_estimator=[int(x) for x in np.linspace (start = 100, stop = 1200, num = 12)]

# Number of features to consider at every split

max_features = ['auto', 'sqrt']
# Maximum number of Levels in tree

max_depth = [int(x) for x in np.linspace (5, 30, num = 6)]

#max_depth.append(None)
# Minimum number of samples required to split a node

min_samples_split = [2, 5, 10, 15, 100]

#Minimum number of samples required at each Leaf node

min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split, 
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune

rf=RandomForestRegressor()
```

```python
rf_random=RandomizedSearchCV(estimator=rf, 
                             param_distributions = random_grid,
                             scoring='neg_mean_squared_error',
                             n_iter =10,
                             cv =5,
                             verbose=2,
                             random_state=42,
                             n_jobs=1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
predictions =rf_random.predict(X_test)
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```
